```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the stock symbol for a Finnish company
# For example, Nokia Corporation is listed on the Helsinki Stock Exchange
stock_symbol = 'NESTE.HE'  # '.HE' denotes Helsinki Exchange

# Define the date range for historical data
start_date = '2001-01-01'
end_date = '2024-09-02'

# Fetch historical data for the Finnish stock
finnish_stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
```

```python
# Display the fetched data
print(finnish_stock_data.head())
```

```python
# Plot the entire time series
plt.figure(figsize=(10, 6))
plt.plot(finnish_stock_data['Close'], label='Stock Price')
plt.title('Historical Stock Price for Nokia Corporation')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the path and file name for the Excel file
file_path = r'C:\Users\Ricks\OneDrive\Skrivbord\Aktie data\finnish_stock_data.xlsx'

# Export the data to an Excel file
finnish_stock_data.to_excel(file_path, index=True)  # Set index=True to include the DataFrame index in the Excel file
```
